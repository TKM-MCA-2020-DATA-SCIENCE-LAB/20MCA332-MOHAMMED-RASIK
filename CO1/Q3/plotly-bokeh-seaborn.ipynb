```python
import seaborn as s
dataset=s.load_dataset('tips')
dataset.head(10)
```

```python
s.distplot(dataset['tip'])
```

```python
s.distplot(dataset['total_bill'],kde=False)
```

```python
s.distplot(dataset['total_bill'],kde=False,bins=130)
```

```python
s.jointplot(x='total_bill',y='tip',data=dataset,kind='scatter')
```

```python
s.jointplot(x='total_bill',y='tip',data=dataset,kind='hex') 
```

```python
s.jointplot(x='total_bill',y='tip',data=dataset,kind='reg')
```

```python

s.jointplot(x='total_bill',y='tip',data=dataset,kind='resid')
```

```python
s.pairplot(dataset)
```

```python
s.pairplot(dataset,hue='sex',palette='coolwarm')
```

```python
s.rugplot(dataset['tip'])
```

## CATEGORICAL PLOTS

```python
dataset.head(6)
```

```python
s.barplot(x='sex',y='total_bill',data=dataset)
```

```python
import numpy as np
s.barplot(x='sex',y='total_bill',data=dataset,estimator=np.std)
```

```python
# COUNTPLOT here y is implicitly set as count of x co ordinate

s.countplot(x='sex',data=dataset)
```

```python
#BOXPLOT
s.boxplot(x='day',y='total_bill',data=dataset,palette='rainbow')
```

```python
#HORIZONTAL

s.boxplot(data=dataset,palette='coolwarm',orient='h')
```

```python
s.boxplot(x='day',y='total_bill',data=dataset,palette='rainbow',hue='smoker')
```

```python
#VIOLINPLOT

s.violinplot(x='day',y='total_bill',data=dataset,palette='rainbow',hue='sex')
```

```python
#COMPARISON

s.violinplot(x='day',y='total_bill',data=dataset,palette='rainbow',hue='sex',split=True)
```

```python
#STRIP PLOT

s.stripplot(x='day',y='total_bill',data=dataset)
```

```python
s.stripplot(x='day',y='total_bill',data=dataset,jitter=True)
```

```python
s.stripplot(x='day',y='total_bill',data=dataset,jitter=True,hue='sex',palette='coolwarm')
```

```python
# FACTOR PLOT

s.factorplot(x='sex',y='total_bill',data=dataset,kind='bar')
```

```python
s.catplot(x='sex',y='total_bill',data=dataset,kind='bar')
```
