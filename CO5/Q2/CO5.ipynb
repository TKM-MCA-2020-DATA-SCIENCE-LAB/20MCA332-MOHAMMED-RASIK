```python
from bs4 import BeautifulSoup
import csv
import requests
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
webpage=requests.get("https://www.icc-cricket.com/rankings/mens/player-rankings/test/batting")
```

```python
soup=BeautifulSoup(webpage.content,"html.parser")
print(soup.prettify())
```

```python
rating=soup.find_all(class_='table-body__cell rating')
```
